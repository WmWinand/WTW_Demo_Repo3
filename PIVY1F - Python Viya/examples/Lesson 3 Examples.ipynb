## 3.2 Analyzing CAS Tables

### Applying Parameters to a CAS Table

#### Using OOP (Object Oriented Programming)

```python
tbl = conn.CASTable('cars', caslib='casuser')
tbl.where = 'Origin = "Asia"'
tbl.vars = ['Make', 'Model', 'MSRP']

tbl.fetch(to=5)
```

```python
## Using params level
tbl = conn.CASTable('cars', caslib='casuser')
tbl.params.where = 'Origin = "Asia"'
tbl.params.vars = ['Make', 'Model', 'MSRP']

tbl.fetch(to=5)
```

#### Deleting Parameters

```python
## Deleting parameters

# Method 1
tbl.del_params('where', 'vars')

# Method 2
#del tbl.where
#del tbl.vars
```

#### Statically Typed

```python
conn.fetch(table = {'name' : 'cars',
                    'caslib' : 'casuser',
                    'where' : 'Origin = "Asia"',
                    'vars' : ['Make', 'Model', 'MSRP']},
           to = 5)
```

```python
inputTbl = dict(name = 'cars', 
                caslib = 'casuser', 
                where = 'Origin = "Asia"',
                vars = ['Make', 'Model', 'MSRP'])

conn.fetch(table = inputTbl, to = 5)
```

#### Using the freqTab.freqTab action

```python
conn.loadActionSet('freqTab')

cars.freqTab(tabulate = [
                'Origin',
                'Cylinders',
                {'vars':['Origin','Cylinders']}
            ]
)
```

#### Using the simple.crossTab action

```python
cars_crossdf = cars.crossTab(row='Origin', 
                             col='DriveTrain', 
                             aggregator='MEAN', 
                             weight='MSRP')['Crosstab']
display(cars_crossdf)


## Rename columns using SASDataFrame labels
cars_crossdf.apply_labels(inplace = True)
display(cars_crossdf)
```

## 3.3 Preparing CAS Tables

### Using the CasOut Parameter in Actions

#### Without the casOut parameter

```python
cars = conn.CASTable('cars', caslib='casuser')
cars.vars = ['Make', 'MSRP', 'MPG_City', 'MPG_Highway']

cars.summary(subset=['Mean', 'Min', 'Max'])
```

#### Using the casOut parameter

```python
cars = conn.CASTable('cars', caslib = 'casuser')
cars.vars = ['Make', 'MSRP', 'MPG_City', 'MPG_Highway']

cars.summary(subset = ['Mean', 'Min', 'Max'],
             casOut = {'name':'cars_summary',
                       'caslib':'casuser'})
```

```python
conn.tableInfo(caslib = 'casuser')
```

```python
cars_summary = conn.CASTable('cars_summary')
cars_summary.head()
```

### Executing a SAS DATA Step

```python
conn.runCode(
   code=
   '''
   data casuser.Asia
        casuser.Europe
        casuser.USA;
      set casuser.cars;
      if Origin='Asia' then output casuser.Asia;
        else if Origin='Europe' then output casuser.Europe;
        else if Origin='USA' then output casuser.USA;
   run;
   ''')    
```

```python
conn.tableInfo(caslib = 'casuser')
```

## 3.4 Executing SQL in CAS

### Executing FedSQL in CAS

```python
conn.loadActionSet('fedSQL')
conn.fedSQL?
```

#### Simple query

```python
conn.execDirect(query="""select Make, Model, MSRP 
                           from casuser.cars 
                           limit 10""")
```

#### Create a table

```python
conn.execDirect(query="""create table casuser.cars_10 as 
                         select Make, Model, MSRP 
                           from casuser.cars 
                           limit 10""")
```

```python
conn.tableInfo(caslib = 'casuser')
```

#### Save a table as a data source file

```python
conn.save(table = {'name':'cars_10', 'caslib':'casuser'},  ## CAS Table to save
          name = 'cars.csv', caslib = 'casuser')           ## New file information
```

```python
conn.fileInfo(caslib = 'casuser')
```

#### Delete the data source file

```python
conn.deleteSource(source = 'cars.csv', caslib = 'casuser', quiet = True)
```

```python
conn.fileInfo(caslib = 'casuser')
```

#### Promote a session scope table

```python
conn.tableInfo(caslib = 'casuser', name = 'cars_10')
```

```python
conn.promote(name = 'cars_10', caslib = 'casuser',       ## input CAS table
             target = 'cars_10', targetlib = 'casuser')  ## promoted CAS table
```

```python
conn.tableInfo(caslib = 'casuser', name = 'cars_10')
```

#### Drop the table

```python
conn.dropTable(name = 'cars_10', caslib = 'casuser', quiet = True)
```
